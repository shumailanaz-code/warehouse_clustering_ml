```python
# Import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.environ["OMP_NUM_THREADS"] = '1'
```

```python
# Load dataset

wh_data = pd.read_csv("Warehouse_data.csv")
```

```python
# Filling the missig values of the dataset, as missing values are in numercial cloumn we will 'mean' to fill it

wh_data = wh_data.fillna(wh_data.Quantity_Redelivered.mean())
wh_data.head()
```

```python
# Checking the data again for missing values

wh_data.isnull().sum()
```

```python
# Convert 'Quantity_Redelivered' column into integer

wh_data.Quantity_Redelivered = wh_data.Quantity_Redelivered.astype(int)
wh_data.head()
```

```python
sns.pairplot(wh_data, vars=['Quantity_Ordered', 'Quantity_Delivered', 'Quantity_Returned', 'Quantity_Redelivered'])
plt.suptitle('Scatter Plots of Quantity Ordered, Delivered, Returned and Redelivered', y=1.05)
plt.show()
```

```python
# Input Varibales: we will use 'Quantity Ordered', 'Quantity Delivered' Quantity Returned' and 'Quantity Redelivered as input

X = wh_data.iloc[:,[6,7,8,9]].values
X[0:3]
```

```python
# Data Normalization

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_SC = scaler.fit_transform(X)
X_SC[0:3]
```

```python
# Step 2: Application of Hierarchical Clustering- Agglomerative Method
```

```python
# (a) Computation of number of clusters by using Dendrogram

import scipy.cluster.hierarchy as sch

plt.figure(figsize=(9,6))
dend = sch.dendrogram(sch.linkage(X_SC, method = 'ward'))
plt.title('Dendrogram of Orders Status')
plt.show()
```

```python
# (b) Application of Hierarchical Clustering- Agglomerative Method

from sklearn.cluster import AgglomerativeClustering

clusters = AgglomerativeClustering(n_clusters = 3, metric = 'euclidean', linkage = 'ward')
y_pred_hc = clusters.fit_predict(X_SC)
```

```python
# View of the dataset with labels after Kmeans application

wh_data['Order_Status'] = y_pred_hc
wh_data.head()
```

```python
# Scatter plot of 'Quantity Ordered' and 'Quantity Delivered' to view clusters

plt.figure(figsize=(6,6))

plt.scatter(X_SC[y_pred_hc == 0, 0], X_SC[y_pred_hc == 0, 1], c = 'blue', label = 'Clsuter1')
plt.scatter(X_SC[y_pred_hc == 1, 0], X_SC[y_pred_hc == 1, 1], c = 'green', label = 'Clsuter2')
plt.scatter(X_SC[y_pred_hc == 2, 0], X_SC[y_pred_hc == 2, 1], c = 'purple', label = 'Clsuter3')

plt.title('Clusters of Orders Status')
plt.xlabel('Quantity Ordered')
plt.ylabel('Quantity Delivered')
plt.legend()
plt.show()
```

```python
# Scatter plot of 'Quantity Ordered' and 'Quantity Returned' to view clusters with centroids

plt.figure(figsize=(6,6))

plt.scatter(X_SC[y_pred_hc == 0, 0], X_SC[y_pred_hc == 0, 2], c = 'blue', label = 'Clsuter1')
plt.scatter(X_SC[y_pred_hc == 1, 0], X_SC[y_pred_hc == 1, 2], c = 'green', label = 'Clsuter2')
plt.scatter(X_SC[y_pred_hc == 2, 0], X_SC[y_pred_hc == 2, 2], c = 'purple', label = 'Clsuter3')

plt.title('Clusters of Orders Status')
plt.xlabel('Quantity Ordered')
plt.ylabel('Quantity Returned')
plt.legend()
plt.show()
```

```python
# Scatter plot of 'Quantity Ordered' and 'Quantity Redelivered' to view clusters

plt.figure(figsize=(6,6))

plt.scatter(X_SC[y_pred_hc == 0, 0], X_SC[y_pred_hc == 0, 3], c = 'blue', label = 'Clsuter1')
plt.scatter(X_SC[y_pred_hc == 1, 0], X_SC[y_pred_hc == 1, 3], c = 'green', label = 'Clsuter2')
plt.scatter(X_SC[y_pred_hc == 2, 0], X_SC[y_pred_hc == 2, 3], c = 'purple', label = 'Clsuter3')

plt.title('Clusters of Orders Status')
plt.xlabel('Quantity Ordered')
plt.ylabel('Quantity Redelivered')
plt.legend()
plt.show()
```
